```
# 以下に動画ファイルのパスを張り付ける
%env FILENAME=/content/batting.mp4

# detectron2のインストール
!pip install detectron2 -f https://dl.fbaipublicfiles.com/detectron2/wheels/cu101/index.html

!pip install -U pyyaml

# VideoPose3DwithDetectron2のダウンロード
!git clone https://github.com/johndpope/VideoPose3DwithDetectron2

# 学習済みモデルのダウンロード
%cd VideoPose3DwithDetectron2/detectron2/detectron2-infer
!wget https://dl.fbaipublicfiles.com/detectron2/COCO-Keypoints/keypoint_rcnn_X_101_32x8d_FPN_3x/139686956/model_final_5ad38f.pkl

%cd /content/VideoPose3DwithDetectron2
!mkdir checkpoint
%cd checkpoint
!wget https://dl.fbaipublicfiles.com/video-pose-3d/pretrained_h36m_detectron_coco.bin

# データの準備
%cd /content/VideoPose3DwithDetectron2/data/
!python prepare_data_2d_custom.py -i /content -o myvideos

# 姿勢推定処理
%cd /content/VideoPose3DwithDetectron2/detectron2/detectron2-infer
!python detectron_pose_predictor.py $FILENAME out

# 動画の生成
%cd /content/VideoPose3DwithDetectron2/
!python run.py -d custom -k myvideos -arc 3,3,3,3,3 -c checkpoint --evaluate pretrained_h36m_detectron_coco.bin --render --viz-subject detectron2 --viz-action custom --viz-camera 0 --viz-video $FILENAME --viz-output /content/out.mp4 --viz-size 6
```
